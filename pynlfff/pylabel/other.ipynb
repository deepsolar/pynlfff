```python
SELECT harpnum_trec, h48_flare_level, h48_flare_id, t_rec_datetime, a.deeps_flare_id , a.start_datetime ,a.end_datetime ,a.peak_datetime ,a.xray_class ,a.xray_intensity 
into public.a202203_flare48
FROM public.a202203_flare a, public.a202203_nlfff_flare  b
where a.deeps_flare_id = b.h48_flare_id;
```

```python
ALTER TABLE public.a202203_flare48 ADD xray_class_value int4 NULL;
ALTER TABLE public.a202203_flare48 ADD xray_intensity_value int4 NULL;


update public.a202203_flare48 set xray_class_value =1 where  xray_class ='A';
update public.a202203_flare48 set xray_class_value =10 where  xray_class ='B';
update public.a202203_flare48 set xray_class_value =100 where  xray_class ='C';
update public.a202203_flare48 set xray_class_value =1000 where  xray_class ='M';
update public.a202203_flare48 set xray_class_value =10000 where  xray_class ='X';
update public.a202203_flare48 set xray_intensity_value  = xray_class_value * xray_intensity ;

```

```python
SELECT harpnum_trec, h48_flare_level, t_rec_datetime, deeps_flare_id, start_datetime, end_datetime, peak_datetime, 
start_datetime::TIMESTAMP-t_rec_datetime::TIMESTAMP as nowflarestart,
end_datetime::TIMESTAMP-t_rec_datetime::TIMESTAMP as nowflareend , 
peak_datetime::TIMESTAMP-t_rec_datetime::TIMESTAMP as nowflarepeak , 
extract(epoch FROM (start_datetime - t_rec_datetime )) as numnowflarestart,
extract(epoch FROM (end_datetime - t_rec_datetime )) as numnowflareend,
extract(epoch FROM (peak_datetime - t_rec_datetime )) as numnowflarepeak,
xray_intensity, xray_class_value, xray_intensity_value
into public.a202203_flare48_data2
FROM public.a202203_flare48;


```
